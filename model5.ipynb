```python
%matplotlib inline

import pandas as pd
import numpy as np
import logging
import warnings
import os
import json

from prophet import Prophet
from prophet.plot import add_changepoints_to_plot
from prophet.serialize import model_to_json
import matplotlib.pyplot as plt

warnings.filterwarnings('ignore')
logging.getLogger('cmdstanpy').setLevel(logging.WARNING)

# =============================================================
# 1. FUNGSI STATUS GIZI (BAHASA RAMAH IBU-IBU)
# =============================================================
def klasifikasi_status_gizi_permenkes(z_score, indikator):
    if indikator == 'BB/U':
        if z_score < -3: return "Berat Badan Sangat Kurang (Perlu Konsultasi Bidan/Dokter)"
        elif -3 <= z_score < -2: return "Beresiko Berat Badan Kurang (Ayo Tingkatkan Nutrisi!)"
        elif -2 <= z_score <= 1: return "Berat Badan Normal (Pertumbuhan Aman)"
        else: return "Beresiko Berat Badan Lebih (Waspada Kegemukan)"
        
    elif indikator == 'TB/U':
        if z_score < -3: return "Sangat Pendek (Perlu Perhatian Khusus)"
        elif -3 <= z_score < -2: return "Beresiko Pendek (Waspada Indikasi Stunting)"
        elif -2 <= z_score <= 3: return "Tinggi Badan Normal (Pertumbuhan Aman)"
        else: return "Cenderung Lebih Tinggi dari Umurnya"
        
    elif indikator == 'LK/U':
        if z_score < -2: return "Lingkar Kepala Lebih Kecil dari Normal"
        elif -2 <= z_score <= 2: return "Lingkar Kepala Normal (Aman)"
        else: return "Lingkar Kepala Lebih Besar dari Normal"
        
    return "Tidak Diketahui"

# =============================================================
# 2. LOAD DATASET
# =============================================================
print("="*60)
print("MEMUAT DATASET...")
print("="*60)

# Pastikan path ini sesuai
df = pd.read_excel(r'D:\SKRIPSI\Sistem_posyandu\DATA_FINAL_1_POSYANDU_FIX.xlsx')
df.columns = df.columns.str.strip().str.lower()
df['tanggal_ukur'] = pd.to_datetime(df['tanggal_ukur'])
df = df.dropna(subset=['nama_anak', 'tanggal_ukur'])
df = df.sort_values(['nama_anak', 'tanggal_ukur'])
df = df.drop_duplicates()

print(f"Total data: {len(df)}")

# =============================================================
# 3. SETUP & PERSIAPAN
# =============================================================
daftar_anak = df['nama_anak'].unique()
daftar_indikator = ['berat', 'tinggi', 'lingkar_kepala']

# FOLDER BARU SESUAI PERMINTAAN
nama_folder_export = 'exported_models_1'
os.makedirs(nama_folder_export, exist_ok=True)
metadata = {}

# =============================================================
# 4. LOOP PER INDIKATOR
# =============================================================
for metrik in daftar_indikator:
    print("\n" + "="*80)
    print(f"ANALISIS INDIKATOR : {metrik.upper()}")
    print("="*80)
    
    hasil_evaluasi_global = []

    for nama in daftar_anak:
        df_anak = df[df['nama_anak'] == nama].copy()
        df_anak = df_anak.sort_values('tanggal_ukur')

        if metrik not in df_anak.columns: continue

        df_prophet = df_anak[['tanggal_ukur', metrik]].copy()
        df_prophet = df_prophet.rename(columns={'tanggal_ukur': 'ds', metrik: 'y'}).dropna()
        df_prophet = df_prophet[df_prophet['y'] > 0]

        # Filter Anomali Ekstrem Biologis
        if metrik == 'berat': df_prophet = df_prophet[(df_prophet['y'] >= 2) & (df_prophet['y'] <= 40)]
        elif metrik == 'tinggi': df_prophet = df_prophet[(df_prophet['y'] >= 30) & (df_prophet['y'] <= 130)]
        elif metrik == 'lingkar_kepala': df_prophet = df_prophet[(df_prophet['y'] >= 20) & (df_prophet['y'] <= 60)]

        df_prophet = df_prophet.drop_duplicates(subset=['ds']).sort_values('ds')

        # Limit Batas Atas (Cap)
        if metrik == 'berat': df_prophet['cap'] = 35
        elif metrik == 'tinggi': df_prophet['cap'] = 130
        else: df_prophet['cap'] = 60

        if len(df_prophet) < 10 or df_prophet['y'].nunique() < 2: continue

        # --- EVALUASI MODEL (TRAIN 80% / TEST 20%) ---
        train_size = int(len(df_prophet) * 0.8)
        data_latih = df_prophet.iloc[:train_size]
        data_uji = df_prophet.iloc[train_size:]

        if len(data_uji) < 2: continue

        # Parameter Model: Logistic, Interval 95%, Sangat Stabil (0.005)
        model_eval = Prophet(
            growth='logistic',
            interval_width=0.95,  
            yearly_seasonality=False,
            weekly_seasonality=False,
            daily_seasonality=False,
            changepoint_prior_scale=0.005,
            changepoint_range=0.8
        )

        try:
            model_eval.fit(data_latih)
        except Exception: continue

        df_uji_pred = data_uji[['ds']].copy()
        df_uji_pred['cap'] = data_latih['cap'].iloc[0]
        forecast_eval = model_eval.predict(df_uji_pred)
        
        y_true = data_uji['y'].values
        y_pred = forecast_eval['yhat'].values

        error = y_true - y_pred
        mae = np.mean(np.abs(error))
        rmse = np.sqrt(np.mean(error**2))
        mape = np.mean(np.abs(error / np.where(y_true == 0, 1, y_true))) * 100

        hasil_evaluasi_global.append({'Nama': nama, 'MAE': mae, 'RMSE': rmse, 'MAPE': mape})

        # --- EXPORT MODEL FINAL PER ANAK (TRAIN 100%) ---
        model_final = Prophet(
            growth='logistic', interval_width=0.95, yearly_seasonality=False,
            weekly_seasonality=False, daily_seasonality=False,
            changepoint_prior_scale=0.005, changepoint_range=0.8
        )
        model_final.fit(df_prophet)
        
        nama_file_aman = nama.replace(" ", "_").replace("/", "_")
        model_path = f'{nama_folder_export}/prophet_{metrik}_{nama_file_aman}.json'
        with open(model_path, 'w') as f:
            json.dump(model_to_json(model_final), f)

    # =========================================================
    # 5. HASIL EVALUASI GLOBAL
    # =========================================================
    if len(hasil_evaluasi_global) == 0:
        print("Tidak ada data valid")
        continue

    df_eval = pd.DataFrame(hasil_evaluasi_global)
    rata_mae, rata_rmse, rata_mape = df_eval['MAE'].mean(), df_eval['RMSE'].mean(), df_eval['MAPE'].mean()
    akurasi = 100 - rata_mape
    satuan = "Kg" if metrik == 'berat' else "Cm"

    print("\nHASIL GLOBAL")
    print("-"*50)
    print(f"Jumlah Anak Dievaluasi : {len(df_eval)}\nMAE                  : {rata_mae:.2f} {satuan}\nRMSE                 : {rata_rmse:.2f} {satuan}\nMAPE                 : {rata_mape:.2f}%\nAKURASI RATA-RATA    : {akurasi:.2f}%")

    # =========================================================
    # 6. STUDI KASUS (Visualisasi Terbaik, Menengah, Terburuk)
    # =========================================================
    idx_terbaik = df_eval['MAPE'].idxmin()
    idx_terburuk = df_eval['MAPE'].idxmax()
    df_eval['selisih_mean'] = abs(df_eval['MAPE'] - rata_mape)
    idx_menengah = df_eval['selisih_mean'].idxmin()

    studi_kasus = {
        'TERBAIK': df_eval.loc[idx_terbaik, 'Nama'],
        'MENENGAH': df_eval.loc[idx_menengah, 'Nama'],
        'TERBURUK': df_eval.loc[idx_terburuk, 'Nama']
    }

    kode_ind = 'BB/U' if metrik == 'berat' else 'TB/U' if metrik == 'tinggi' else 'LK/U'

    for kategori, nama_target in studi_kasus.items():
        print("\n" + "="*70)
        print(f"STUDI KASUS {kategori}")
        print("="*70)

        eval_anak = df_eval[df_eval['Nama'] == nama_target].iloc[0]
        df_anak = df[df['nama_anak'] == nama_target].copy().sort_values('tanggal_ukur')
        df_prophet = df_anak[['tanggal_ukur', metrik]].copy().rename(columns={'tanggal_ukur': 'ds', metrik: 'y'}).dropna()
        
        if metrik == 'berat': df_prophet['cap'] = 35
        elif metrik == 'tinggi': df_prophet['cap'] = 130
        else: df_prophet['cap'] = 60

        train_size = int(len(df_prophet) * 0.8)
        data_latih = df_prophet.iloc[:train_size]
        data_uji = df_prophet.iloc[train_size:]

        model = Prophet(
            growth='logistic', 
            interval_width=0.95, 
            yearly_seasonality=False,
            weekly_seasonality=False, 
            daily_seasonality=False,
            changepoint_prior_scale=0.005, 
            changepoint_range=0.8
        )
        model.fit(data_latih)

        df_uji_pred = data_uji[['ds']].copy()
        df_uji_pred['cap'] = data_latih['cap'].iloc[0]
        forecast = model.predict(df_uji_pred)

        # PENDEKATAN Z-SCORE STATISTIK (Hanya untuk Visualisasi, tidak perlu tanggal lahir)
        mean_aktual = data_latih['y'].mean()
        std_aktual = data_latih['y'].std()
        prediksi_terakhir = forecast['yhat'].iloc[-1]
        
        if std_aktual == 0: z_score = 0
        else: z_score = round((prediksi_terakhir - mean_aktual) / std_aktual, 2)
        
        z_score = max(-4, min(4, z_score)) 
        status_gizi = klasifikasi_status_gizi_permenkes(z_score, kode_ind)

        print(f"Nama Anak          : {nama_target}")
        print(f"MAPE               : {eval_anak['MAPE']:.2f}%")
        print(f"Akurasi Prediksi   : {100 - eval_anak['MAPE']:.2f}%")
        print(f"Z-Score Pendekatan : {z_score}")
        print(f"Status             : {status_gizi}")

        hasil = pd.DataFrame({
            'Tanggal': data_uji['ds'].dt.strftime('%Y-%m-%d'),
            'Aktual': np.round(data_uji['y'].values, 2),
            'Prediksi': np.round(forecast['yhat'].values, 2)
        })
        hasil['Error'] = np.round(np.abs(hasil['Aktual'] - hasil['Prediksi']), 2)

        print("\nTABEL PREDIKSI")
        print(hasil.to_string(index=False))

        # =========================================================
        # GRAFIK 1: Default Prophet (Sumbu Diperlebar)
        # =========================================================
        fig1 = model.plot(forecast, figsize=(14, 6)) 
        add_changepoints_to_plot(fig1.gca(), model, forecast)
        plt.title(f'Tren Prophet {metrik.upper()} - {nama_target}', fontweight='bold')
        plt.grid(True)
        plt.show()

        # =========================================================
        # GRAFIK 2: Aktual vs Prediksi Data Uji (Sumbu Diperlebar)
        # =========================================================
        plt.figure(figsize=(14, 6)) 
        plt.plot(data_uji['ds'], data_uji['y'], marker='o', color='blue', linewidth=2, label='Aktual (Data Uji)')
        plt.plot(data_uji['ds'], forecast['yhat'], marker='x', linestyle='--', color='red', linewidth=2, label='Prediksi Prophet')
        plt.fill_between(data_uji['ds'], forecast['yhat_lower'], forecast['yhat_upper'], color='red', alpha=0.2, label='Confidence Interval 95%')
        plt.title(f'Perbandingan Aktual vs Prediksi {metrik.upper()} - {nama_target}', fontweight='bold')
        plt.xlabel('Tanggal Pengukuran')
        plt.ylabel(f'{metrik.upper()} ({satuan})')
        plt.legend()
        plt.grid(True)
        plt.show()

    # METADATA
    metadata[metrik] = {
        "jumlah_anak_diekspor": int(len(df_eval)),
        "akurasi_rata_rata": round(float(akurasi), 3),
        "mae_rata_rata": round(float(rata_mae), 3),
        "rmse_rata_rata": round(float(rata_rmse), 3),
        "mape_rata_rata": round(float(rata_mape), 3),
        "satuan": satuan
    }

# =============================================================
# 7. SIMPAN METADATA
# =============================================================
metadata_path = f'{nama_folder_export}/metadata_sistem.json'
with open(metadata_path, 'w') as f:
    json.dump(metadata, f, indent=4)

print("\n" + "="*60)
print(f"PROSES SELESAI: SEMUA MODEL BERHASIL DIEKSPOR KE FOLDER '{nama_folder_export}'")
print("="*60)
```
